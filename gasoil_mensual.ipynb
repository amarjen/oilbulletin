```python
import glob
filenames = glob.glob("./raw_data/*")
```

```python
df = pd.concat([pd.read_excel(file) for file in filenames], ignore_index=True)
```

```python
df2 = df[(df["Country Name"]=="Spain") & (df["Product Name"]=="Automotive gas oil")][["Prices in force on", "Weekly price with taxes"]]
df2.columns = ["Fecha", "Precio"]
df2 = df2.set_index("Fecha").sort_index(ascending=False)
df2['Precio'] = df2.apply(lambda x: float(str(x['Precio']).replace(",","")), axis = 1)
```
